```python
# Cargamos los datos: leemos el csv y obtenemos un dataframe
df_nba_games = pd.read_csv('/home/dsc/repos/master-kschool/TFM/games.csv')
df_nba_games
```

```python
# Nos quedamos solo con los valores numericos
df_nba_games_only_numerics = df_nba_games.select_dtypes(include=np.number)
df_nba_games_only_numerics
```

```python
# Comprobamos cuantos huecos en blanco tenemos
```

```python
# Obtengo las columnas con celdas vacías
columns_with_nan = df_nba_games_only_numerics.columns[df_nba_games_only_numerics.isnull().any()]
print(columns_with_nan)
```

```python
# Cuantos valores NaN hay para una columna:
for column in columns_with_nan:
    count_nan = df_nba_games_only_numerics[column].isna().sum()
    print ('Column: ' + str(column) + ' has :' + str(count_nan) + ' of NaN')
```

```python
# Es decir tenemos 99 Filas sin algun resultado -> vamos a identificar esas filas
```

```python
# Obtenemos un dataframe fila y cantidad de nan:
df_fila_and_nans = df_nba_games_only_numerics.isnull().sum(axis=1)
df_fila_and_nans

#Creamos una lista vacia donde meteremos las filas que contienen valores nan:
filas_conflictivas = []

for i in range(0, 25023):
    if df_fila_and_nans[i] != 0:# si es distinto de cero quiere decir que algun nan tiene
        filas_conflictivas.append(i)# en caso afirmativo añadimos la fila a la lista de "filas conflictivas"
        
filas_conflictivas
len(filas_conflictivas)

# Tenemos guardadas en filas_conflictivas las filas con valores nulos
# Hemos comprobado que tenemos 99 filas afectadas, 
#que es lo mismo que obteniamos mirando la cantidad de nulos para cada columna
```
